## 2. Data Cleaning

A. Dibagian data cleaning kita akan memanipulasi beberapa data meliputi penggantian nama kolom, penghapusan kolom, dan penambahan kolom. 

```python
# Kolom yang mengandung simbol yang ingin dihapus
columns_to_modify = df.select_dtypes(include='object').columns

# Menghapus simbol '$', ',', dan '-'
df[columns_to_modify] = df[columns_to_modify].replace('[\$,\\-,]', '', regex=True)
df.columns
```

```python
# ganti nama kolom supaya mempermudah pengecekan data
column_Change = {
    'Discount Band': 'Discount_Band',
    'Units Sold': 'Units_Sold',
    'Manufacturing Price': 'Manufacturing_Price',
    'Sale Price': 'Sale_Price',
    'Gross Sales': 'Gross_Sales',
    'Month Number': 'Month_Number',
    'Month Name': 'Month_Name',
    'Date' : 'Tanggal',
    'Year' : 'Tahun'
}
df.rename(columns=column_Change, inplace=True)
df.columns
```

```python
#menghapus spasi yang ada di dalam dataset
def replace(n):
    n = n.str.strip()
    return n

HapusSpasi=[]
for n in df.columns.to_list():
    if df[n].dtype==object:
        HapusSpasi.append(n)

df[HapusSpasi] = df[HapusSpasi].apply(replace)
df
```

B. Penggantian tipe data yang sesuai dengan seharusnya. 

```python
#code berikut berfungsi untuk mengganti data type kolom tertentu. 
columns_to_convert = ['Units_Sold', 'Manufacturing_Price', 'Sale_Price', 'Gross_Sales', 'Discounts', 'Sales', 'COGS']
df[columns_to_convert] = df[columns_to_convert].astype(str).apply(lambda x: x.str.strip()).replace('', '0').astype(float)
df.info()
```

```python
# mengganti data type Date object menjadi datetime
df['Tanggal'] = pd.to_datetime(df['Tanggal'])
```

```python
# Cek dan handling missing value dengan imputasi angka nol '0'
data= []
for profit in df.Profit:
    profit = profit.replace(" ","")
    try:
        if "(" in profit:
            negativeValue = float(profit.strip('()'))
            data.append(negativeValue)
        else:
            data.append(float(profit))
    except:
        data.append(float(0))

df.Profit = data
df
```

```python
# code segment_mapping berfungsi sebagai dictionary keys and value 
country_mapping = {
    'Canada': 1,
    'Germany': 2,
    'France': 3,
    'Mexico': 4,
    'United States of America': 5
}

# membuat kolom baru 'country_ID' dengan numerical values berdasarkan dictionary
df['country_ID'] = df['Country'].replace(country_mapping)

# menampikan values dati country_ID dan Country 
df[['country_ID','Country']].drop_duplicates()
```
