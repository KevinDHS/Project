## C.Analisis dan perhitungan

* Statistik Deskriptif

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import scipy.stats as stats
from scipy.stats import f_oneway

# Hitung rata-rata gaji
mean_salary = data['salary_in_usd'].mean()

# Hitung median gaji
median_salary = data['salary_in_usd'].median()

# Hitung mode gaji
mode_salary = data['salary_in_usd'].mode()

# Tampilkan hasil
print("Rata-rata gaji: ", mean_salary)
print("Median gaji: ", median_salary)
print("Modes gaji: ", mode_salary)
```

* Statistik Inferensial

didalam statistik inferensial saya akan menggunakan perhitungan confidence interval.

```python
# Mengelompokkan data berdasarkan level pengalaman
junior_salaries = data[data['experience_level'] == 'Entry-Level']['salary_in_usd']
mid_salaries = data[data['experience_level'] == 'Mid-Level']['salary_in_usd']
senior_salaries = data[data['experience_level'] == 'Senior']['salary_in_usd']
executive_salaries = data[data['experience_level'] == 'Executive']['salary_in_usd']

# Menghitung mean dan standard deviation untuk setiap kelompok
mean_junior = np.mean(junior_salaries)
std_junior = np.std(junior_salaries)
mean_mid = np.mean(mid_salaries)
std_mid = np.std(mid_salaries)
mean_senior = np.mean(senior_salaries)
std_senior = np.std(senior_salaries)
mean_executive = np.mean(executive_salaries)
std_exevutive = np.std(executive_salaries)

# Menghitung confidence interval untuk masing-masing kelompok dengan tingkat kepercayaan 95%
conf_int_junior = stats.norm.interval(0.95, loc=mean_junior, scale=std_junior/np.sqrt(len(junior_salaries)))
conf_int_mid = stats.norm.interval(0.95, loc=mean_mid, scale=std_mid/np.sqrt(len(mid_salaries)))
conf_int_senior = stats.norm.interval(0.95, loc=mean_senior, scale=std_senior/np.sqrt(len(senior_salaries)))
conf_int_executive = stats.norm.interval(0.95, loc=mean_executive, scale=std_senior/np.sqrt(len(executive_salaries)))

print("Confidence interval untuk rata-rata gaji Entry-Level':", conf_int_junior)
print("Confidence interval untuk rata-rata gaji Mid-Level:", conf_int_mid)
print("Confidence interval untuk rata-rata gaji Senior:", conf_int_senior)
print("Confidence interval untuk rata-rata gaji Executive:", conf_int_executive)
```

Dari hasil confidence interval untuk rata-rata gaji pada setiap level pengalaman yang Anda berikan, kita dapat membuat beberapa kesimpulan:

1. **Entry-Level**: Confidence interval untuk rata-rata gaji pada level entry adalah antara $85,857.25 dan $94,510.54. Artinya, dengan tingkat kepercayaan 95%, rata-rata gaji untuk pekerja pada level entry dapat diperkirakan berada dalam kisaran tersebut.

2. **Mid-Level**: Confidence interval untuk rata-rata gaji pada level mid adalah antara $116,458.83 dan $121,481.86. Ini menunjukkan bahwa rata-rata gaji untuk pekerja pada level mid berada dalam kisaran tersebut dengan tingkat kepercayaan 95%.

3. **Senior**: Confidence interval untuk rata-rata gaji pada level senior adalah antara $161,270.73 dan $164,108.71. Hal ini menunjukkan bahwa rata-rata gaji untuk pekerja pada level senior berada dalam kisaran tersebut dengan tingkat kepercayaan 95%.

4. **Executive**: Confidence interval untuk rata-rata gaji pada level executive adalah antara $183,407.08 dan $197,274.14. Ini menunjukkan bahwa rata-rata gaji untuk pekerja pada level eksekutif berada dalam kisaran tersebut dengan tingkat kepercayaan 95%.

Dengan demikian, kita dapat menyimpulkan bahwa, dengan tingkat kepercayaan 95%, kita memiliki keyakinan bahwa rata-rata gaji pada setiap level pengalaman berada dalam kisaran confidence interval yang diberikan. Perbedaan antara level pengalaman juga dapat dilihat dari perbedaan rentang confidence interval yang diberikan.

* Hypothesis testing

* saya menggunakan fungsi `f_oneway` dari modul `scipy.stats` untuk melakukan uji `ANOVA`. Hasil dari uji ini akan memberikan nilai `F-statistic` dan `p-value`. Nilai p-value yang lebih rendah dari tingkat signifikansi yang ditetapkan (biasanya 0,05) akan menunjukkan adanya perbedaan yang signifikan antara rata-rata gaji di antara kelompok level pengalaman.

- H0: Tidak ada perbedaan signifikan dalam rata-rata gaji antara kelompok level pengalaman (Entry-Level, Mid-Level, Senior, Executive).

- H1: Ada perbedaan signifikan dalam rata-rata gaji antara setidaknya dua kelompok level pengalaman.

```python
junior_salaries = data[data['experience_level'] == 'Entry-Level']['salary_in_usd']
mid_salaries = data[data['experience_level'] == 'Mid-Level']['salary_in_usd']
senior_salaries = data[data['experience_level'] == 'Senior']['salary_in_usd']
executive_salaries = data[data['experience_level'] == 'Executive']['salary_in_usd']

# Melakukan uji ANOVA
f_statistic, p_value = f_oneway(junior_salaries, mid_salaries, senior_salaries, executive_salaries)

print("Nilai F-statistic:", f_statistic)
print("Nilai p-value:", p_value)
```

Dalam kasus ini, nilai F-statistic adalah 516.3816 dan nilai p-value sangat kecil, mendekati nol . Dengan nilai p-value yang sangat kecil, kita memiliki bukti yang kuat untuk menolak hipotesis nol. Artinya, ada perbedaan yang signifikan antara rata-rata gaji di antara setidaknya dua kelompok level pengalaman (Entry-Level, Mid-Level, Senior, dan Executive).

* visualisasi data



pada bagian visualisasi data ini pertama saya akan membuat kondisi antara `upper_salary` dan `lower_salary`

```python
# Menghitung rata-rata salary
average_salary = data['salary_in_usd'].mean()

# Menentukan threshold untuk upper_salary dan lower_salary (misal: 10% dari rata-rata)
threshold = 0.1 * average_salary

# Menghitung upper_salary, average_salary, dan lower_salary
upper_salary = average_salary + threshold
lower_salary = average_salary - threshold

# Menambahkan kolom 'salary_category' berdasarkan kondisi
data['salary_category'] = pd.cut(data['salary_in_usd'], bins=[-np.inf, lower_salary, upper_salary, np.inf], labels=['below_average', 'average', 'above_average'])

# Plotting
plt.figure(figsize=(10, 6))
sns.countplot(x='salary_category', data=data, order=['below_average', 'average', 'above_average'])
plt.title('Distribusi Kategori Gaji')
plt.xlabel('Salary Category')
plt.ylabel('Count Total')
plt.show()
```

dari tabel informasi bar chart diatas dapat dilihat bahwa, distribusi salary di bagian `below_average` lebih besar. 

1. Apakah jenis employee juga berpengaruh pada besar salary yang diterima? dan termasuk upper_salary,average_salary,lower_salary

pada bagian visualisasi ini, saya akan melakukan pencarian berdasrkan `upper,lower,average` berdasarkan jenis employee.

```python
import numpy as np

# Menghitung rata-rata gaji untuk setiap jenis employee
average_salary_by_category = data.groupby('job_category')['salary_in_usd'].mean()

# Menentukan persentase untuk batasan "upper" dan "lower"
upper_percentage = 1.1  # 10% di atas rata-rata
lower_percentage = 0.9  # 10% di bawah rata-rata

# Menghitung upper dan lower salary berdasarkan rata-rata gaji
upper_salary_by_category = average_salary_by_category * upper_percentage
lower_salary_by_category = average_salary_by_category * lower_percentage

# Menentukan kategori employee berdasarkan gaji mereka
def categorize_employee_salary(row):
    if row['salary_in_usd'] > upper_salary_by_category[row['job_category']]:
        return 'upper'
    elif row['salary_in_usd'] < lower_salary_by_category[row['job_category']]:
        return 'lower'
    else:
        return 'average'

# Menambahkan kolom 'salary_category' berdasarkan kondisi
data['salary_category'] = data.apply(categorize_employee_salary, axis=1)

# Plotting
plt.figure(figsize=(12, 6))
sns.countplot(x='job_category', hue='salary_category', data=data, order=data['job_category'].value_counts().index)
plt.title('Distribusi Kategori Gaji berdasarkan Jenis Employee')
plt.xlabel('Job Category')
plt.ylabel('Count Total')
plt.xticks(rotation=90)
plt.legend(title='Salary Category')
plt.show()
```

dari bar chart diatas dapat disimpulkan bahwa rata2 employee mendapat salary di `below_average`. 

2. Apakah lokasi company berdampak pada perbedaan jumlah salary?

selanjutnya saya akan mencari perbedaan salary berdasarkan lokasi company. dengan memfilter top 5 saja.

```python
# Melakukan grouping dan perhitungan rata-rata salary berdasarkan lokasi employee
average_salary_by_location = data.groupby('company_location')['salary_in_usd'].mean().sort_values(ascending=False).head(5)

# Plotting
plt.figure(figsize=(10, 6))
sns.barplot(x=average_salary_by_location.index, y=average_salary_by_location.values)
plt.title('Perbedaan Jumlah Salaries Berdasarkan Company Lokasi  (Top 5)')
plt.xlabel('Company Location')
plt.ylabel('Average Salary (USD)')
plt.xticks(rotation=45)
plt.show()
```

dari data diatas dapat disimpulkan bahwa rata2 pembagian salary terjadi gap perubahan yang cukup signifikan di Qatal dari beberapa negara.

3. Job Title apa yang memiliki jumlah salary tertinggi dan terendah?

selanjutnya saya akan melihat job title apa saja yang memiliki jumlah salary tertinggi dan terendah, dengan menampilkan percentage. 

```python
job_title_highest_salary = data[data['salary_in_usd'] == data['salary_in_usd'].max()]['job_title'].iloc[0]
job_title_lowest_salary = data[data['salary_in_usd'] == data['salary_in_usd'].min()]['job_title'].iloc[0]

job_titles = [job_title_highest_salary, job_title_lowest_salary]
salaries = [data['salary_in_usd'].max(), data['salary_in_usd'].min()]

# Plotting
plt.figure(figsize=(8, 8))
plt.pie(salaries, labels=job_titles, autopct='%1.1f%%', colors=['blue', 'red'])
plt.title('Persentase Salary Tertinggi dan Terendah berdasarkan Job Title')
plt.show()
```
